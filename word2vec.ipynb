```python
import pandas as pd
```

```python
import glob
text = "\n".join([open(file, 'r',encoding='utf-8').read() for file in glob.glob('txts/*.txt')])
```

```python
print(text)
```

```python
from nltk.tokenize import  word_tokenize
import nltk
nltk.download('punkt')
```

```python
tokens = word_tokenize(text.lower())
```

```python
import re
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.isalnum() and word not in stop_words]
filtered_tokens = [re.sub(r'\d+', '', word) for word in filtered_tokens]
# filtered_tokens= [re.sub(r'[^a-zA-Z\s]', '', word) for word in filtered_tokens]
```

```python
from gensim.models import Word2Vec



w2v_model = Word2Vec([filtered_tokens], window=5, min_count=2)

```

```python
w2v_model.wv.most_similar('vision', topn=10)
```

```python
w2v_model.wv.doesnt_match(['tony','banner', 'steve','barton', 'natasha', 'clint', 'thor','groot'])
```

```python
w2v_model.wv.get_normed_vectors().shape
```

```python
w2v_model.wv.similarity('tony', 'captain')
```
